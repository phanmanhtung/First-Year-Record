```python
#########

# Create a function to store a subject result as tuple

def subj(name, score_20, ECTs):
    score_4 = 0
    weighted_score = 0
    
    # convert to 4.0 scale:
    if score_20 >= 10 and score_20 < 11:
        score_4 = 2.67
    elif score_20 >= 11 and score_20 < 13:
        score_4 = 3.00
    elif score_20 >= 13 and score_20 < 14:
        score_4 = 3.33
    elif score_20 >= 14 and score_20 < 16:
        score_4 = 3.67
    elif score_20 >= 16:
        score_4 = 4.00
        
    weighted_score = score_4 * ECTs
        
    return (name, score_20, score_4, ECTs, weighted_score)
```
